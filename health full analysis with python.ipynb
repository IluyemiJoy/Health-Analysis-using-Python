# Creating Categories for the Numerical Columns 

```python
# For age
def med_age_groups(Age):
    if Age < 20:
        return 'Adolescents'
    elif Age < 40:
        return 'Young Adults'
    elif Age < 60:
        return 'Middle Aged Adults'
    elif Age < 80:
        return 'Seniors'
    else:
        return 'Elderly'


health_dataset['Age group'] = health_dataset['Age'].apply(med_age_groups)

health_dataset
```

```python
#For Years of Smoking (0-39)

def years_smoking(years):
    if years == 0:
        return 'Never'
    elif years <= 5:
        return '1-5 years'
    elif years <= 10:
        return '6-10 years'
    elif years <= 20:
        return '11-20 years'
    elif years <= 30:
        return '21-30 years'
    else:
        return '31+ years'

health_dataset['Smoking Duration']= health_dataset['Years_of_Smoking'].apply(years_smoking)

health_dataset
```

```python
# For Cigarettes per day
def cig_category(cigs):
    if cigs == 0:
        return 'Non Smoker'
    elif cigs <= 5:
        return '1-5 cigs/day'
    elif cigs <= 10:
        return '6-10 cigs/day'
    elif cigs <= 15:
        return '11-15 cigs/day'
    elif cigs <= 20:
        return '16-20 cigs/day'
    elif cigs <= 25:
        return '21-25 cigs/day'
    else:
        return '26+ cigs/day'

health_dataset['Cigarette Intensity']= health_dataset['Cigarettes_Per_Day'].apply(cig_category)

health_dataset
```

```python
# For BMI ( based on WHO categorization)
def bmi_category(bmi):
    if pd.isna(bmi):
        return 'Missing'
    elif bmi < 18.5:
        return 'Underweight'
    elif bmi < 25:
        return 'Normal Weight'
    elif bmi < 30:
        return 'Overweight'
    elif bmi < 35:
        return 'Obesity Class I'
    elif bmi < 40:
        return 'Obesity Class II'
    else:
        return 'Obesity Class III'

health_dataset['BMI Category']= health_dataset['BMI'].apply(bmi_category)

health_dataset
```

```python
# For Cholesterol
def cholesterol_category(cholesterol):
    if pd.isna(cholesterol):
        return 'Missing'
    elif cholesterol < 200:
        return 'Desirable'
    elif cholesterol < 240:
        return 'High'
    else:
        return 'Very High'

health_dataset['Cholesterol Category']= health_dataset['Cholesterol_Level'].apply(cholesterol_category)

health_dataset
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# Create distributions
smoking_distribution = health_dataset.groupby(['Smoking Duration', 'Smoking_Status', 'Organ_Condition']).agg(
    Patient_Count=('Patient_ID', 'count'),
).round(2).reset_index()

BMI_distribution = health_dataset.groupby(['BMI Category', 'Organ_Condition']).agg(
    Patient_Count=('Patient_ID', 'count'),
).round(2).reset_index()

# Get average BMI by smoking duration and organ condition
Avg_BMI_by_condition = health_dataset.groupby(['Smoking Duration', 'Organ_Condition'])['BMI'].mean().round(2).reset_index()

# Get unique organ conditions
organ_conditions = sorted(BMI_distribution['Organ_Condition'].unique().tolist())

# Create initial figure
fig = make_subplots(
    rows=2, cols=2,
    subplot_titles=('Smoking Duration Analysis', 
                    'BMI Category Distribution',
                    'Average BMI by Smoking Duration',
                    'Condition Summary'),
    specs=[[{'type': 'bar'}, {'type': 'bar'}],
           [{'type': 'scatter'}, {'type': 'table'}]]
)

# Create dropdown buttons
buttons = []

for i, condition in enumerate(organ_conditions):
    visibility = [False] * (len(organ_conditions) * 3 + 1)
    visibility[i*3] = True
    visibility[i*3 + 1] = True
    visibility[i*3 + 2] = True
    visibility[-1] = True
    
    buttons.append(
        dict(
            label=condition,
            method="update",
            args=[{"visible": visibility},
                  {"title": f"Organ Condition: {condition}"}]
        )
    )

# Add "Show All" button
all_visible = [True] * (len(organ_conditions) * 3 + 1)
buttons.insert(0, 
    dict(label="All Conditions",
         method="update",
         args=[{"visible": all_visible},
               {"title": "All Organ Conditions"}])
)

# Add traces
colors = {'Damaged': 'red', 'Healthy': 'green'}
smoking_order = ['Never', '1-5 years', '6-10 years', '11-20 years', '21-30 years', '31+ years']
bmi_order = ['Underweight', 'Normal', 'Overweight', 'Obese']

for condition in organ_conditions:
    # Chart 1: Smoking Duration
    smoking_data = smoking_distribution[smoking_distribution['Organ_Condition'] == condition]
    smoking_total = smoking_data.groupby('Smoking Duration')['Patient_Count'].sum().reset_index()
    
    # Ensure correct order
    smoking_map = {dur: i for i, dur in enumerate(smoking_order)}
    smoking_total['order'] = smoking_total['Smoking Duration'].map(smoking_map)
    smoking_total = smoking_total.sort_values('order')
    
    fig.add_trace(
        go.Bar(x=smoking_total['Smoking Duration'], 
               y=smoking_total['Patient_Count'],
               name=condition,
               marker_color=colors.get(condition, 'blue'),
               visible=False),
        row=1, col=1
    )
    
    # Chart 2: BMI Category
    bmi_data = BMI_distribution[BMI_distribution['Organ_Condition'] == condition]
    
    # Ensure correct order
    bmi_map = {cat: i for i, cat in enumerate(bmi_order)}
    bmi_data['order'] = bmi_data['BMI Category'].map(bmi_map)
    bmi_data = bmi_data.sort_values('order')
    
    fig.add_trace(
        go.Bar(x=bmi_data['BMI Category'], 
               y=bmi_data['Patient_Count'],
               name=condition,
               marker_color=colors.get(condition, 'blue'),
               visible=False),
        row=1, col=2
    )
    
    # Chart 3: Average BMI
    bmi_avg_data = Avg_BMI_by_condition[Avg_BMI_by_condition['Organ_Condition'] == condition]
    
    # Ensure correct order
    bmi_avg_data['order'] = bmi_avg_data['Smoking Duration'].map(smoking_map)
    bmi_avg_data = bmi_avg_data.sort_values('order')
    
    fig.add_trace(
        go.Scatter(x=bmi_avg_data['Smoking Duration'], 
                  y=bmi_avg_data['BMI'],
                  mode='lines+markers',
                  line=dict(width=3),
                  marker=dict(size=10, color=colors.get(condition, 'blue')),
                  name=condition,
                  visible=False),
        row=2, col=1
    )

# Add summary table (SIMPLIFIED - no average smoking duration)
summary_data = []
for condition in organ_conditions:
    condition_data = health_dataset[health_dataset['Organ_Condition'] == condition]
    
    total_patients = len(condition_data)
    avg_bmi = condition_data['BMI'].mean()
    
    summary_data.append([
        condition,
        int(total_patients),
        f"{avg_bmi:.1f}"
    ])

fig.add_trace(
    go.Table(
        header=dict(values=['Organ Condition', 'Total Patients', 'Average BMI'],
                    fill_color='lightblue',
                    align='center',
                    font=dict(size=12)),
        cells=dict(values=[list(zip(*summary_data))[0] if summary_data else [],
                          list(zip(*summary_data))[1] if summary_data else [],
                          list(zip(*summary_data))[2] if summary_data else []],
                   fill_color='lavender',
                   align='center',
                   font=dict(size=11)),
        visible=True
    ),
    row=2, col=2
)

# Update layout
fig.update_layout(
    updatemenus=[dict(
        buttons=buttons,
        direction="down",
        pad={"r": 10, "t": 10},
        showactive=True,
        x=0.1,
        xanchor="left",
        y=1.15,
        yanchor="top"
    )],
    height=800,
    width=1200,
    title_text="Health Risk Analysis by Organ Condition",
    showlegend=True,
    barmode='group'
)

# Update axis labels
fig.update_xaxes(title_text="Smoking Duration", row=1, col=1, categoryorder='array', categoryarray=smoking_order)
fig.update_yaxes(title_text="Patient Count", row=1, col=1)
fig.update_xaxes(title_text="BMI Category", row=1, col=2, categoryorder='array', categoryarray=bmi_order)
fig.update_yaxes(title_text="Patient Count", row=1, col=2)
fig.update_xaxes(title_text="Smoking Duration", row=2, col=1, categoryorder='array', categoryarray=smoking_order)
fig.update_yaxes(title_text="Average BMI", row=2, col=1)

fig.show()
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import pandas as pd

# Create distributions (without gender split - just total counts by condition)
fam_distribution = health_dataset.groupby(['Family_History_Risk', 'Organ_Condition'])['Patient_ID'].count().reset_index(name='Patient_Count')
bp_distribution = health_dataset.groupby(['BP_Risk', 'Organ_Condition'])['Patient_ID'].count().reset_index(name='Patient_Count')
Cholesterol_distribution = health_dataset.groupby(['Cholesterol Category', 'Organ_Condition'])['Patient_ID'].count().reset_index(name='Patient_Count')
Cig_distribution = health_dataset.groupby(['Cigarette Intensity', 'Organ_Condition'])['Patient_ID'].count().reset_index(name='Patient_Count')
alcohol_distribution = health_dataset.groupby(['Alcohol_Consumption', 'Organ_Condition'])['Patient_ID'].count().reset_index(name='Patient_Count')

# Get unique organ conditions and genders
organ_conditions = sorted(health_dataset['Organ_Condition'].unique().tolist())
genders = sorted(health_dataset['Gender'].unique().tolist())

# Create initial figure with 3x2 layout
fig = make_subplots(
    rows=3, cols=2,
    subplot_titles=('Family History Risk', 
                    'Blood Pressure Risk',
                    'Cholesterol Category',
                    'Cigarette Intensity',
                    'Alcohol Consumption',
                    'Risk Factors Summary'),
    specs=[[{'type': 'bar'}, {'type': 'bar'}],
           [{'type': 'bar'}, {'type': 'bar'}],
           [{'type': 'bar'}, {'type': 'table'}]],
    vertical_spacing=0.1,
    horizontal_spacing=0.15
)

# Create DOUBLE DROPDOWN: Organ Condition + Gender
buttons = []

# First, create organ condition buttons
for i, condition in enumerate(organ_conditions):
    for j, gender in enumerate(genders):
        # Create visibility array for each combination
        visibility = [False] * (len(organ_conditions) * len(genders) * 5 + 1)
        
        # Calculate index for this combination
        combo_index = i * len(genders) + j
        
        # Set visibility for all 5 charts for this combination
        for chart_num in range(5):
            visibility[combo_index * 5 + chart_num] = True
        
        visibility[-1] = True  # Table always visible
        
        buttons.append(
            dict(
                label=f"{condition} - {gender}",
                method="update",
                args=[{"visible": visibility},
                      {"title": f"Risk Factors: {condition} ({gender})"}]
            )
        )

# Add "Show All" button for all conditions and genders
all_visible = [True] * (len(organ_conditions) * len(genders) * 5 + 1)
buttons.insert(0, 
    dict(label="All Conditions & Genders",
         method="update",
         args=[{"visible": all_visible},
               {"title": "Risk Factors: All Data"}])
)

# Colors for conditions
colors = {'Damaged': 'red', 'Healthy': 'green'}

# Add traces for each organ condition AND gender combination
for i, condition in enumerate(organ_conditions):
    for j, gender in enumerate(genders):
        # Filter data for this condition AND gender
        condition_gender_data = health_dataset[(health_dataset['Organ_Condition'] == condition) & 
                                               (health_dataset['Gender'] == gender)]
        
        # Chart 1: Family History Risk
        fam_data = condition_gender_data.groupby('Family_History_Risk').size().reset_index(name='Count')
        fig.add_trace(
            go.Bar(
                x=fam_data['Family_History_Risk'],
                y=fam_data['Count'],
                name=f"{condition} ({gender})",
                marker_color=colors.get(condition, 'blue'),
                legendgroup=f"{condition}_{gender}",
                showlegend=True if i == 0 and j == 0 else False,  # Show only first legend
                visible=False
            ),
            row=1, col=1
        )
        
        # Chart 2: Blood Pressure Risk
        bp_data = condition_gender_data.groupby('BP_Risk').size().reset_index(name='Count')
        fig.add_trace(
            go.Bar(
                x=bp_data['BP_Risk'],
                y=bp_data['Count'],
                name=f"{condition} ({gender})",
                marker_color=colors.get(condition, 'blue'),
                legendgroup=f"{condition}_{gender}",
                showlegend=False,
                visible=False
            ),
            row=1, col=2
        )
        
        # Chart 3: Cholesterol Category
        chol_data = condition_gender_data.groupby('Cholesterol Category').size().reset_index(name='Count')
        fig.add_trace(
            go.Bar(
                x=chol_data['Cholesterol Category'],
                y=chol_data['Count'],
                name=f"{condition} ({gender})",
                marker_color=colors.get(condition, 'blue'),
                legendgroup=f"{condition}_{gender}",
                showlegend=False,
                visible=False
            ),
            row=2, col=1
        )
        
        # Chart 4: Cigarette Intensity
        cig_data = condition_gender_data.groupby('Cigarette Intensity').size().reset_index(name='Count')
        fig.add_trace(
            go.Bar(
                x=cig_data['Cigarette Intensity'],
                y=cig_data['Count'],
                name=f"{condition} ({gender})",
                marker_color=colors.get(condition, 'blue'),
                legendgroup=f"{condition}_{gender}",
                showlegend=False,
                visible=False
            ),
            row=2, col=2
        )
        
        # Chart 5: Alcohol Consumption
        alcohol_data = condition_gender_data.groupby('Alcohol_Consumption').size().reset_index(name='Count')
        fig.add_trace(
            go.Bar(
                x=alcohol_data['Alcohol_Consumption'],
                y=alcohol_data['Count'],
                name=f"{condition} ({gender})",
                marker_color=colors.get(condition, 'blue'),
                legendgroup=f"{condition}_{gender}",
                showlegend=False,
                visible=False
            ),
            row=3, col=1
        )

# Create summary table (aggregated across genders)
summary_data = []
for condition in organ_conditions:
    condition_data = health_dataset[health_dataset['Organ_Condition'] == condition]
    
    total_patients = len(condition_data)
    
    # Calculate risk percentages
    fam_high = (condition_data['Family_History_Risk'] == 'High').mean() * 100
    bp_high = (condition_data['BP_Risk'] == 'High').mean() * 100
    chol_high = (condition_data['Cholesterol Category'] == 'High').mean() * 100
    
    # Cigarette usage
    cig_smoker = (condition_data['Cigarette Intensity'] != 'Non-smoker').mean() * 100
    
    # Alcohol consumption
    alcohol_drinker = (condition_data['Alcohol_Consumption'] != 'Non-drinker').mean() * 100
    
    summary_data.append([
        condition,
        int(total_patients),
        f"{fam_high:.1f}%",
        f"{bp_high:.1f}%",
        f"{chol_high:.1f}%",
        f"{cig_smoker:.1f}%",
        f"{alcohol_drinker:.1f}%"
    ])

fig.add_trace(
    go.Table(
        header=dict(values=['Condition', 'Total Patients', 'High Family History', 
                           'High BP', 'High Cholesterol', 'Smokers %', 'Alcohol Drinkers %'],
                    fill_color='lightblue',
                    align='center',
                    font=dict(size=10)),
        cells=dict(values=[list(zip(*summary_data))[0] if summary_data else [],
                          list(zip(*summary_data))[1] if summary_data else [],
                          list(zip(*summary_data))[2] if summary_data else [],
                          list(zip(*summary_data))[3] if summary_data else [],
                          list(zip(*summary_data))[4] if summary_data else [],
                          list(zip(*summary_data))[5] if summary_data else [],
                          list(zip(*summary_data))[6] if summary_data else []],
                   fill_color='lavender',
                   align='center',
                   font=dict(size=9)),
        visible=True
    ),
    row=3, col=2
)

# Update layout
fig.update_layout(
    updatemenus=[dict(
        buttons=buttons,
        direction="down",
        pad={"r": 10, "t": 10},
        showactive=True,
        x=0.1,
        xanchor="left",
        y=1.15,
        yanchor="top"
    )],
    height=1000,
    width=1400,
    title_text="Risk Factors Analysis by Organ Condition & Gender",
    showlegend=True,
    barmode='group',
    template="plotly_white",
    legend=dict(
        orientation="h",
        yanchor="bottom",
        y=1.02,
        xanchor="right",
        x=1
    )
)

# Update axis labels
fig.update_xaxes(title_text="Family History Risk", row=1, col=1)
fig.update_yaxes(title_text="Patient Count", row=1, col=1)
fig.update_xaxes(title_text="Blood Pressure Risk", row=1, col=2)
fig.update_yaxes(title_text="Patient Count", row=1, col=2)
fig.update_xaxes(title_text="Cholesterol Category", row=2, col=1)
fig.update_yaxes(title_text="Patient Count", row=2, col=1)
fig.update_xaxes(title_text="Cigarette Intensity", row=2, col=2)
fig.update_yaxes(title_text="Patient Count", row=2, col=2)
fig.update_xaxes(title_text="Alcohol Consumption", row=3, col=1)
fig.update_yaxes(title_text="Patient Count", row=3, col=1)

fig.show()
```
